# Plotting extra examples

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from sklearn.mixture import GaussianMixture
```

## Plotting the mixture model

Using a mixture model with 4 components, the probability density will be a combination of 4 normal distributions, which may be univariate (one dimensional) or multivariate (2 or more dimensions) depending on the dimensions of the data.

In this example I have generated 100 data points in 2 dimensions

```python
K = 4 # 4 mixture components
d = np.random.uniform(size=(100,2)) # 100 random points in two dimensions
mixturemodel = GaussianMixture(K) 
mixturemodel.fit(d)
```

To visualise the results, as we have a two dimensional data, we can plot the mixture model density in 2d using contours, that give us an idea of the shape of the density. 

The number of components is independent of the number of dimensions, so we can have a mixture of K two dimensional multivariate normal distributions.

The code generates a range of x and y locations, and takes all pairs of of (x,y) coordinates and calculates the density of the mixture model at that location. This uses the parameters of the fitted mixture model, which has K components, each having a weight, mean (a two dimensional vector), and covariance matrix.

```python
xs = np.linspace(-0.5,1.5,200) # Range of values on x axis
ys = np.linspace(-0.5,1.5,200) # Range of values on y axis

zs = np.zeros((200,200)) # z values -- height of the contour plot at each (x,y) location

for i,x in enumerate(xs):
    for j,y in enumerate(ys):
        z = 0.0
        # For each mixture component sum up the density of the 2d multivariate normal at (x,y)
        for k in range(K):
            z += mixturemodel.weights_[k]*multivariate_normal(mixturemodel.means_[k], mixturemodel.covariances_[k]).pdf([x,y])
        zs[j,i] = z
        
# Plot the data points
plt.scatter(d[:,0],d[:,1],c=mixturemodel.predict(d))              
# Add red crosses for the means of each component
plt.scatter(mixturemodel.means_[:,0],mixturemodel.means_[:,1],marker="x",c='r')
# Plot contours of the mixture density 
plt.contour(xs, ys, zs)
```

### More than two dimensional data

We can also do this with data in more than two dimensions. To visualise the results in two dimensions, we can simply look at the components of the multivariate normal distribution in those dimensions.

In this example we have $P=5$ dimensions in the data and display a plot that slices through dimensions $u=2$ and $v=4$.

In numpy we can use the notation 
        
        a[[u,v]]
    
 to take indices u and v from vector a, and
 
         m[[u,v],[u,v]]
         
 to take rows u and v and columns u and v from matrix m, giving a two by two matrix.

```python
K = 3 # 3 mixture components
P = 5 # 5 dimensional data (5 variables)
d = np.random.uniform(size=(100,P)) # 100 random points in P dimensions
mixturemodel = GaussianMixture(K) 
mixturemodel.fit(d)
```

```python
u = 2 # Plot data for columns d[:,3] and d[:,7]
v = 4

xs = np.linspace(-0.5,1.5,200) # Range of values on x axis
ys = np.linspace(-0.5,1.5,200) # Range of values on y axis

zs = np.zeros((200,200)) # z values -- height of the contour plot at each (x,y) location

for i,x in enumerate(xs):
    for j,y in enumerate(ys):
        z = 0.0
        # For each mixture component sum up the density of the P dimensional multivariate normal at (x,y) in dimensions i and j
        for k in range(K):
            z += mixturemodel.weights_[k]*multivariate_normal(mixturemodel.means_[k][[u,v]], mixturemodel.covariances_[k][[u,v],[u,v]]).pdf([x,y])
        zs[j,i] = z
        
# Plot the data points
plt.scatter(d[:,u],d[:,v],c=mixturemodel.predict(d))              
# Add red crosses for the means of each component
plt.scatter(mixturemodel.means_[:,u],mixturemodel.means_[:,v],marker="x",c='r')
# Plot contours of the mixture density 
plt.contour(xs, ys, zs)
```

## Adding labels for plots

To add a label to a variable on a matplotlib plot, we can use the label argument to specify the label. The we also need to use plt.legend() to tell matplotlib to display the legend giving a key for each label. matplotlib may place the legend in odd places to try and avoid overlapping the plot, don't worry about this for the coursework!

```python
# Generate some random x and y values
xs = np.linspace(0,10,100)
y1 = xs*np.random.uniform(size=(100,))
y2 = xs*np.random.uniform(size=(100,))
y3 = xs*np.random.uniform(size=(100,))
# Plot each different variable as a line plot with a label
plt.plot(xs,y1,label="Label for y1")
plt.plot(xs,y2,label="Label for y2")
plt.plot(xs,y3,label="Label for y3")
# Show a legend giving a key for the different lines
plt.legend()
```

## Log scales

In some plots you may find one variable is much larger, or varies over a larger scale (perhaps 10-10000) than another which varies on a smaller scale (e.g. 1-10).

This can make it very hard to see any variation in the smaller variable:

```python
# Generate some random x and y values
xs = np.linspace(1,10,100)
y1 = xs*np.random.uniform(size=(100,))
y2 = 1000*xs*np.random.uniform(size=(100,))
# Plot without using the logarithm of y
plt.plot(xs,y1,label="Label for y1")
plt.plot(xs,y2,label="Label for y2")
# Show a legend giving a key for the different lines
plt.legend()
```

In matplotlib we can use the plotting functions plt.semilogx, plt.semilogy or plt.loglog to draw the x, y, or both axes on a logarithmic scale.

This has the effect of squashing variables that vary over a larger range.

```python
# Generate some random x and y values
xs = np.linspace(1,10,100)
y1 = xs*np.random.uniform(size=(100,))
y2 = 1000*xs*np.random.uniform(size=(100,))
# Plot taking the logarithm of y
plt.semilogy(xs,y1,label="Label for y1")
plt.semilogy(xs,y2,label="Label for y2")
# Show a legend giving a key for the different lines
# (the legend goes wherever matplotlib decides looks best)
plt.legend()
```

You can also just apply a logarithm to the values using numpy, with np.log

```python
# Generate some random x and y values
xs = np.linspace(1,10,100)
y1 = xs*np.random.uniform(size=(100,))
y2 = 1000*xs*np.random.uniform(size=(100,))
# Plot taking the logarithm of y
plt.plot(xs,np.log(y1),label="log y1")
plt.plot(xs,np.log(y2),label="log y2")
# Show a legend giving a key for the different lines
# (the legend goes wherever matplotlib decides looks best)
plt.legend()
```
